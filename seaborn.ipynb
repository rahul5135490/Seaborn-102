```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from datetime import datetime
```

```python
df_titanic=sns.load_dataset("titanic")
df_iris=sns.load_dataset("iris")
```

```python
df_titanic.columns
```

```python
df_titanic.head()
```

```python
# countplot
```

```python
sns.catplot(x="sex",data=df_titanic,kind="count")
plt.show()
```

```python
sns.catplot(x="embark_town",data=df_titanic,kind="count")
plt.show()
```

```python
sns.catplot(x="deck",data=df_titanic,kind="count")
plt.show()
```

```python
sns.catplot(y="deck",data=df_titanic,kind="count")
plt.show()
```

```python
sns.catplot(x="sex",data=df_titanic,kind="count",hue="class")
plt.show()
```

```python
sns.catplot(x="sex",data=df_titanic,kind="count",hue="embark_town")
plt.show()
```

```python
sns.catplot(x="species",data=df_iris,kind="count",hue="sepal_length")
plt.show()
```

```python
sns.catplot(x="species",data=df_iris,kind="count")
plt.show()
```

```python
#quantitative visualizations
```

```python
sns.kdeplot(df_iris["sepal_length"],color="red",label="Sepal Lenght")
sns.kdeplot(df_iris["sepal_width"],color="blue",label="Sepal Width")
plt.legend()
plt.show()
```

```python
sns.histplot(df_iris["sepal_length"],color="red",label="Sepal Lenght")
sns.histplot(df_iris["sepal_width"],color="blue",label="Sepal Width")
plt.legend()
plt.show()
```

```python
sns.regplot(x=df_iris["petal_length"],y=df_iris["sepal_length"])
```
